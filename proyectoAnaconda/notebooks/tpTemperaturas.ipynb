```python
import pandas as pd
import seaborn as sb


def encontrarMaxima(df):
    for indice, valor in df["media"].items():
        if valor == df["media"].max():
            return df["mes"][indice], df["ao"][indice]

def encontrarMinima(df):
    for indice, valor in df["media"].items():
        if valor == df["media"].min():
            return df["mes"][indice], df["ao"][indice]



df1 = pd.read_csv("../data/historico_temperaturas.csv",sep=",",engine="python" )
print(encontrarMaxima(df1))
print(encontrarMinima(df1))
grafico = sb.pairplot(df1,hue="ao",palette="Spectral",height=5)
```

1- Se puede observar que el mes con la mayor temperatura media es diciembre del ao 2013, asi
como el mes con la menor es julio del 1992.

2-A partir del grfico podemos realizar dos observaciones principales. La primera es que a lo largo de
los aos los valores mximos y mnimos anuales se alejaron y con esto aument la media. Adems, se puede
notar que en los aos 2000 hubo una excepcin donde se registraron las mayores
mximas, mnimas y medias.
