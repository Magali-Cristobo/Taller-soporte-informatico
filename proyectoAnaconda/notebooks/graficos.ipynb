#### Sanciones a operadores por empresa

```python
import pandas as pd
import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2
import numpy as np

df1 = pd.read_csv("../data/sanciones-operadores.csv")
df1 = df1.loc[0:5000]
x_values = df1['Empresa'].unique()
nombres_x=["Converia Argentina","Telecab Argentina", "Coop. de La Pampa","Sion","Paralelo 52 \n TV S.A","Telefonia del Sur","Prima","Coseldi","Ver TV","A.M.X. Argentina S.A.","Telecentro","Telefonia Publica \n y Privada","Telmex Argentina","Telefónica Móivles Argentina","Telecom Personal S.A.","Telefonica de Argentina S.A.","Telecom Argentina S.A."]
x_values= np.flip(x_values)
x_values= np.array(nombres_x)
y_values = df1['Empresa'].value_counts().tolist()
y_values.reverse() #esto hace que vaya de menor a mayor, pero no pasa lo mismo en el eje de las x

plt1.bar(x_values,y_values, width=0.9)
# plt1.title('Sanciones operadores por empresa')
plt1.xlabel('Empresas')
plt1.ylabel('Cantidad de sanciones')
plt1.xticks(rotation=90,fontsize=10)
plt1.show()

plt2.plot(x_values,y_values)
# plt2.title('Sanciones operadores por empresa')
plt2.xlabel('Empresas')
plt2.ylabel('Cantidad de sanciones')
plt2.xticks(rotation=90,fontsize=10)

plt2.show()
```

A partir del grafico se podria decir que las empresas con peor atencion al cliente son
ya que son las que poseen mayor cantidadde reclamos hacia sus operadores.


No obstante, de deberia tener en cuenta la cantidad de clientes que posee cada empresa, ya que alguna de las empresas que aparecen con menos
reclamos podrían tener igual calidad de atención, pero poseer menos reclamos porque tienen menos clientes.
También es importante la fecha de los reclamos, ya que interesa saber si la empresa
tuvo esos reclamos en la antiguedad o si continuan en la actualidad.
