#### Sanciones a operadores por reclamos iniciados por los usuarios

```python
import pandas as pd
import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2
import numpy as np

df1 = pd.read_csv("../data/sanciones-operadores.csv")
df1 = df1.loc[0:5000]

#grafico de barras
x_values = df1['Empresa'].unique()
nombres_x=["Converia Argentina","Telecab Argentina", "Coop. de La Pampa","Sion","Paralelo 52 \n TV S.A","Telefonia del Sur","Prima","Coseldi","Ver TV","A.M.X. Argentina S.A.","Telecentro","Telefonia Publica \n y Privada","Telmex Argentina","Telefónica Móivles Argentina","Telecom Personal S.A.","Telefonica de Argentina S.A.","Telecom Argentina S.A."]
x_values= np.flip(x_values)
x_values= np.array(nombres_x)
y_values = df1['Empresa'].value_counts().tolist()
y_values.reverse()
plt1.figure(figsize=(10,6))
plt1.bar(x_values,y_values, width=0.9)
plt1.xlabel('Empresas')
plt1.ylabel('Cantidad de sanciones')
plt1.xticks(rotation=90,fontsize=11)
plt1.show()
```

A partir del gráfico se podría decir que las empresas con peor atención al cliente son Telecom y Telefónica,
ya que son las que poseen mayor cantidad de reclamos hacia sus operadores.

No obstante, se debería tener en cuenta la cantidad de clientes que posee cada empresa, ya que alguna de las empresas que aparecen con menos
reclamos podría tener igual calidad de atención, pero poseer menos reclamos al tener menos clientes.
También es importante la fecha de los reclamos, ya que interesa saber si la empresa
tuvo esos reclamos en la antigüedad o si continuan en la actualidad.


```python
x_values = df1['Discriminación'].unique()
y_values = df1['Discriminación'].value_counts().tolist()
x_values= np.flip(x_values)
y_values.reverse()
plt2.figure(figsize=(10,6))
plt2.plot(x_values,y_values)
plt2.xlabel('Discriminaciones')
plt2.ylabel('Cantidad de sanciones')
plt2.xticks(rotation=90,fontsize=11)

plt2.show()
```

A partir del gráfico se puede decir que las empresas suelen tener problemas al reparar e instalar servicios y
cambiar el domicilio, ya que la mayoría de los reclamos son a raíz de estas tres actividades.
De todos modos, esto no significa que todas las empresas tengan problemas con estas actividades;
podría suceder que una empresa con muchos clientes reciba muchos reclamos sobre estas actividades
y el resto no, pero el gráfico seguiría mostrando que la mayoría de los reclamos a los operadores
son por estos servicios.

Es por esto que va a depender de la empresa, la cantidad de clientes y reclamos que posea,
no se puede afirmar completamente que la mayoría de las empresas tienen problemas con la reparación
e instalación de servicios y cambios de domicilio.
