```python
import pandas as pd
import seaborn as sb

def eliminarPesos(valor):
    valor = str(valor)
    if valor != 'nan':
        valor = valor.replace("$","")
        valor = valor.replace(".","")
        return int(valor)


df1 = pd.read_csv("../data/sanciones-operadores.csv",sep=",",engine="python" )
df1 = df1.loc[0:2500]
df1["Multa (UT)"] = df1["Multa (UT)"].apply(eliminarPesos)
grafico = sb.pairplot(df1,hue="Empresa",palette="Spectral",height=5)
```

A partir de los grficos podemos observar que la empresa
con mayores reclamos es telecom seguida de telefnica y los picos de reclamos
sucedieron en 2017 y 2018. Adems, estas empresas son las que recibieron multas
de mayor monto por lo tanto se puede deducir que recibieron sanciones ms graves, aunque en su mayora las multas
no son elevadas.


Aclaracin: Para realizar los grficos utilizamos las primeras 2000 lineas del csv.
