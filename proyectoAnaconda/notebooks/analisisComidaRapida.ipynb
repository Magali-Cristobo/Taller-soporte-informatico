## Consumo de comida rápida

El trabajo consiste en analizar el consumo de comida rápida o delivery según edad, momento del día, frecuencia, y si piden
en las cadenas de comida rápida más conocidas o no.
Los gráficos que se presentarán a continuación se realizaron a partir de los resultados de esta encuesta: https://docs.google.com/forms/d/e/1FAIpQLSfpEelqHYgjKXK1eo8FWEcz9ljDtNhw0iIX2UKin1x9ipa0Jg/viewform?usp=sf_link ,
en la que participaron a 66 personas, que fueron comunicadas por WhatsApp.

Se realizaron tres gráficos de barras sobre el consumo de comida rápida según edad, cadenas de comida rápida
y frecuencia, y un gráfico de torta para representar el consumo de comida rápida según el momento del día

```python
import pandas as pd
import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2
import matplotlib.pyplot as plt3
import matplotlib.pyplot as plt4
import numpy as np

df1 = pd.read_csv("../data/comidaRapida.csv")
df1 = df1.loc[0:67]
```

```python
x_values = df1['edad'].value_counts().index.tolist()
y_values = df1['edad'].value_counts().tolist()
x_values= np.flip(x_values)
y_values.reverse()
plt1.figure(figsize=(16,10))
plt1.bar(x_values,y_values, width=0.9)
plt1.xticks(x_values,fontsize=11)
plt1.yticks(y_values)
plt1.title('Consumo de comida rápida según edad')
plt1.xlabel('Edad')
plt1.ylabel('Cantidad')
plt1.show()
```

```python
x_values = df1['cantidad de dias'].value_counts().index.tolist()
y_values = df1['cantidad de dias'].value_counts().tolist()
x_values= np.flip(x_values)
y_values.reverse()
plt2.figure(figsize=(16,10))
plt2.bar(x_values,y_values, width=0.9)
plt2.title('Consumo de comida rápida según frecuencia')
plt2.xlabel('Frecuencia')
plt2.ylabel('Cantidad')
plt2.xticks(fontsize=11)
plt2.show()
```

## Conclusión

A partir de los resultados llegué a la conclusión de que, por más de que los resultados son variados, hay más tendencia
a consumir comida rápida por parte de adolescentes (principalmente de 17 a 18 años) y personas mayores a 50 años (52 a 54 años),
aunque no lo hacen muy frecuentemente, una o dos veces por semana y en el momento de la cena. También pude notar que no consumen en las cadenas rápidas más
conocidas como McDonald's, Burger King, Wendy's o Subway, sino que en otras y, a la vez, un gran porcentaje votó que no 
consume comida de cadenas de comida rápida.
