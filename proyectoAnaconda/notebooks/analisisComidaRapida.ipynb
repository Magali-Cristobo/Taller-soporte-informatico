```python
import pandas as pd
import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2
import matplotlib.pyplot as plt3
import matplotlib.pyplot as plt4
import numpy as np

df1 = pd.read_csv("../data/comidaRapida.csv")
df1 = df1.loc[0:67]

#grafico segun edad
diccionario={}
print(df1['edad'].value_counts())
x_values = df1['edad'].value_counts().index.tolist()
y_values = df1['edad'].value_counts().tolist()

for i in range(len(x_values)):
    clave = x_values[i]
    valor = y_values[i]
    lista=[valor]
    diccionario[clave]=lista

df2 = pd.DataFrame(diccionario)
df2.plot(kind="hist")
print(df2)
x_values= np.flip(x_values)
# y_values.reverse()
plt1.figure(figsize=(10,6))
plt1.bar(x_values,y_values, width=0.9)
plt1.title('Consumo de comida rapida segun edad')
plt1.xlabel('Edad')
plt1.ylabel('Cantidad de sanciones')
plt1.xticks(fontsize=11)
plt1.show()

x_values = df1['cantidad de dias'].value_counts().index.tolist()
y_values = df1['cantidad de dias'].value_counts().tolist()
x_values= np.flip(x_values)
y_values.reverse()
plt2.figure(figsize=(10,6))
plt2.bar(x_values,y_values, width=0.9)
plt2.title('Consumo de comida rapida segun frecuencia')
plt2.xlabel('Frecuencia')
plt2.ylabel('Cantidad')
plt2.xticks(fontsize=11)
plt2.show()

x_values = df1['momento del dia'].value_counts().index.tolist()
y_values = df1['momento del dia'].value_counts().tolist()
x_values= np.flip(x_values)
y_values.reverse()
plt3.figure(figsize=(10,6))
plt3.bar(x_values,y_values)
plt3.title('Consumo de comida rapida segun frecuencia')
plt3.xlabel('Frecuencia')
plt3.ylabel('Cantidad')
plt3.xticks(fontsize=11)
plt3.show()

x_values = df1['cadena de comida'].value_counts().index.tolist()
y_values = df1['cadena de comida'].value_counts().tolist()
x_values= np.flip(x_values)
y_values.reverse()
plt4.figure(figsize=(10,6))
plt4.bar(x_values,y_values)
plt4.title('Consumo de comida rapida segun frecuencia')
plt4.xlabel('Frecuencia')
plt4.ylabel('Cantidad')
plt4.xticks(fontsize=11)
plt4.show()
```
